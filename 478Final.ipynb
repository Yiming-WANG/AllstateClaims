```python
#Since the target value "loss" has range from 0.67 to 188318 
#MAE would be a large value. So, would also use MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    #in this project y_true is always > 0
    return np.mean(np.abs((y_true - y_pred) / y_true)) 
```

```python
#put all model performance in "pf" list. This can make it easier to decide which one
#to use in the end.
pf=[]
```

```python
#Since the target of this data set is to predict the 'loss' value, which is 
#180,000 different continuous data. The unsupervised learning algorithms such as
#Kmeans Clustering and Hierarchical Clustering are not appropriate.
#Since this final project require the useage of unsupervised learning algorithms
#would try to use PCA (unsupervised learning) to do data preprocessing.
```

```python
#Will use PCA to catch at least 95% information. It will cut a lot of model 
#computation time. Then compare both of the results. 
#The PAC preprocessed data would only be used for long time computing model
#Such as KNN.
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=60)
X_train_tr_pca = pca.fit(X_train_tr_n).transform(X_train_tr_n)
np.set_printoptions(precision=2,suppress=True)
print(pca.explained_variance_ratio_)
```

#### KNN (Supervised learning)

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Since the data set is real-valued vector spaces, so, compare the performance 
#between  “euclidean”  and “manhattan”. Start with euclidean. Would test K from 
#1 to 10 then store the performance to pf_knn_e and pf_knn_m then add the one 
#with best performance to pf list

#The default value for parameter: "algorithm" is 'auto' that means, it would 
#decide the most appropriate algorithm from {‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}
#based on the values passed to fit method. So, no need to test this parameter.
```

```python
pf_knn_e=[]
for k in range(1,21):
    knn = KNeighborsRegressor(metric='euclidean',n_neighbors=k)
    knn.fit(X_train_tr,Y_train_tr)
    y_knn=np.exp(Y_train_te)
    p_knn=np.exp(knn.predict(X_train_te))
    mae_knn=mean_absolute_error(y_knn, p_knn)
    mape_knn=mean_absolute_percentage_error(y_knn, p_knn)
    pf_knn_e.append([k,mae_knn,mape_knn])
```

```python
for a in pf_knn_e:
    print a
```

```python
pf_knn_m=[]
for k in range(1,21):
    knn = KNeighborsRegressor(metric='manhattan',n_neighbors=k)
    knn.fit(X_train_tr,Y_train_tr)
    y_knn=np.exp(Y_train_te)
    p_knn=np.exp(knn.predict(X_train_te))
    mae_knn=mean_absolute_error(y_knn, p_knn)
    mape_knn=mean_absolute_percentage_error(y_knn, p_knn)
    pf_knn_m.append([k,mae_knn,mape_knn])
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#For the parameter of max_depth. If in default, then nodes are expanded until 
#all leaves are pure or until all leaves contain less than min_samples_split 
#samples. So, would use the default.

#Would test n_estimators from 1 to 20. then store the performance to pf_rf
#then add the one with best performance to pf list
```

```python
pf_rf=[]
for n_estimators in range(1,21):
    rf = RandomForestRegressor(n_estimators=n_estimators)
    rf.fit(X_train_tr_n,Y_train_tr)
    y_rf=np.exp(Y_train_te)
    p_rf=np.exp(rf.predict(X_train_te_n))
    mae_rf=mean_absolute_error(y_rf, p_rf)
    mape_rf=mean_absolute_percentage_error(y_rf, p_rf)
    pf_rf.append([n_estimators,mae_rf,mape_rf])
```

```python
for a in pf_rf:
    print a
```

```python
#The performance is increasing, probably 20 is not best parameter.
#try 10 more:
```

```python
for n_estimators in range(20,31):
    rf = RandomForestRegressor(n_estimators=n_estimators)
    rf.fit(X_train_tr_n,Y_train_tr)
    y_rf=np.exp(Y_train_te)
    p_rf=np.exp(rf.predict(X_train_te_n))
    mae_rf=mean_absolute_error(y_rf, p_rf)
    mape_rf=mean_absolute_percentage_error(y_rf, p_rf)
    pf_rf.append([n_estimators,mae_rf,mape_rf])
```
